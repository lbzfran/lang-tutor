```python
import pandas as pd
import os
```

```python
path = '../data/kr_en/2024_01.csv'
output_path = '../data/kr_en/ke2.json'

df = pd.read_csv(path)
```

```python
df.describe()
```

```python
sdf = df.loc[:,'Korean Definition':'English Definition']
sdf.head()
```

```python
sdf = sdf.rename(columns={'Korean Definition': 'korean', 'English Definition': 'english'})
sdf.head()
```

```python
def clean_list_like_string(list_string):
    list_string = list_string.strip("[]")
    list_string = list_string.replace("'", "").replace('"', "").replace("’", "").replace("‘", "")
    cleaned_string = " ".join(list_string.split(","))
    return cleaned_string
    
sdf['korean'] = sdf['korean'].apply(clean_list_like_string)
sdf['english'] = sdf['english'].apply(clean_list_like_string)
```

```python
sdf.head()
```

```python
sdf.to_json(output_path, force_ascii = False, orient='records')
```
